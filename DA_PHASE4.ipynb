```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/water_potability.csv')
```

```python
print("First 5 rows of the dataset:")
print(data.head())
```

# Histogram

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['Hardness'], bins=30, kde=True)
plt.title('Histogram for Water Quality Analysis')
plt.xlabel('X-Axis Label')
plt.ylabel('Y-Axis Label')
plt.show()
```

# scatter plots

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Solids', y='Hardness', data=data)
plt.title('Scatter Plot for Water Quality Analysis')
plt.xlabel('X-Axis Label')
plt.ylabel('Y-Axis Label')
plt.show()
```

# correlation matrix

```python
correlation_matrix = data.corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix for Water Quality Analysis')
plt.show()
```
